```python
P = tita.groupby("Survived").size().reset_index(name='counts')
P = P.sort_values(['counts'], ascending=True)

P1= list(P.counts)
print(P1)

```

```python
status = ('Survived', 'Fatalities')
y_pos = np.arange(len(status))

numbers = [P1[0] ,P1[1]]
plt.bar(y_pos, numbers, align='center', alpha=0.5, color= ['blue','green'])
plt.xticks(y_pos, status)
plt.ylabel('Count')
plt.title('People of titanic')
plt.show()
```
